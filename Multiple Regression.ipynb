```python
data=pd.read_csv("C:\\Users\\Smita Akolkar\\OneDrive\\Desktop\\Project Data\Home Prices.csv")
data
```

```python
data.isnull()
```

```python
import math
median_bedrooms=math.floor(data.bedrooms.median())
median_bedrooms
```

```python
data.bedrooms=data.bedrooms.fillna(median_bedrooms)
data
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[["area","bedrooms","age"]],data.price)
```

```python
reg.predict([[5000,5,2]])
```
